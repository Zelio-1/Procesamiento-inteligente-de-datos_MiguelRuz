4. Estimar el corto de calefacción para una temperatura externa media de 30°, 5 pulgadas de aislamiento y un calentador con antigüedad

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("SalsberryRealty.csv")
df.head()
```

```python
#Determinar las variables independientes.
x = df[["Temp","Insul","Age"]]
x
```

```python
#Determinar las variable dependiente
y = df[["Cost"]]
y
```

```python
#Ajustar el modelo para encontrar el modelo de regresion múltiple
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(y,y_pred) #Error de la media cuadrática
r2 = r2_score(y,y_pred) #varianza
```

3. Valor de interseccion

```python
model.intercept_
```

```python
#coeficientes
print("Coeficientes son:", model.coef_)
print("Error cuadratico: ", mse)
print("Varianza", r2)
```

4. Estimar parar 30°, aislamiento 5" antiguedad 5 años

```python
Y = model.predict([[30,5,10]])
print("el costo estimado es:", Y)
```
