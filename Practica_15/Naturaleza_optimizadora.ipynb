```python
# Clase Individuo
'''
Proteinas - Mientras mas protesinas mejor. 
Herencia_Limite - Restringe el valor del fenotipo. 
Especie - A que tipo de ser vivo corresponde el individuo.
Ambiente - En donde se desarrolla y está el ser vivo. 
Mutación - Una mutación muy alta podria afectar al fenotipo.
Identificador - Sirve para etiquetar al ser vivo.

Para los puntajes el minimo será 10 y el maximo 100 en general.
'''
class Individuo: 
    def __init__(self, identificador, herencia_limite, proteinas, especie, ambiente, mutacion):
        self.Identificador = identificador
        self.Herencia_Limite = herencia_limite
        self.Proteinas = proteinas
        self.Especie = especie
        self.Ambiente = ambiente
        self.Mutacion = mutacion
    
    def Fenotipo(self): 
        if self.Mutacion > 20: self.Mutacion = self.Mutacion*-1
        result = (self.Proteinas + self.Mutacion) - self.Herencia_Limite
        if result < 0: result = 0 
        return result
    
    def GetEspecie(self):
        return self.Especie
    
    def SetAmbiente(self, value):
        self.Ambiente = value
    
    def GetAmbiente(self):
        return self.Ambiente
    
    def GetIdentificador(self):
        return self.Identificador
```

```python
# Clase Población

class Poblacion():
    def __init__(self, especies, ambiente):
        self.Especies = especies
        self.Ambiente = ambiente
    
    # Top 3 mejores individuos que se adaptan a la población 
    def Seleccion(self, individuos: list[Individuo]): 
        individuos_calificados = []
        for i in range(len(individuos)):
            if individuos[i].GetAmbiente() != self.Ambiente:
                continue
            else:
                fenotipo = individuos[i].Fenotipo()
                if fenotipo >= 50: 
                    individuos_calificados.append(individuos[i].GetIdentificador())
            if len(individuos_calificados) == 3: break
        return individuos_calificados
    
    def GetAmbiente(self):
        return self.Ambiente
```

```python
# Tester

individuos = []

individuo_1 = Individuo("Humano_1", 30, 60, "Humano", "Templado", 43)
pinguino_1 = Individuo("pinguino_1", 10, 87, "Humano", "Artico", 30)
pinguino_2 = Individuo("pinguino_2", 30, 100, "Humano", "Artico", 20)
pinguino_3 = Individuo("pinguino_3", 25, 95, "Humano", "Artico", 18)

individuos.append(individuo_1)
individuos.append(pinguino_1)
individuos.append(pinguino_2)
individuos.append(pinguino_3)

poblacion_1 = Poblacion("Pinguinos", "Artico")

mejores_individuos = poblacion_1.Seleccion(individuos)

print(f'Para la poblacion con un ambiente {poblacion_1.GetAmbiente()} se encontraron que los candidatos mas aptos son: {mejores_individuos}')
```
