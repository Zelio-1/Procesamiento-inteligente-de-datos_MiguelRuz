```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Librerias para operacion de regresion lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('./CopierSalesofAmerica.csv')
```

```python
# Determine la cantidad de filas y columnas del archivo
data.shape
```

```python
# Mostrar la tabla
data.head()
```

```python
# Seleccionar los datos para determinar el coeficiente de correlación
X_train = np.array (data["llamadas"]).reshape(-1, 1)
Y_train = np.array (data["ventas"])
```

```python
# Calcular el coeficiente de correlación lineal
r = LinearRegression() 

# Entrenando al modelo con los datos del .csv
r.fit(X_train, Y_train)

# Obtener las predicciones, valores para y
y_pred = r.predict(X_train)

print('Coeficientes:', r.coef_)
print('Intersección:', r.intercept_)
print('Error cuadrático medio:', mean_squared_error(Y_train, y_pred))
print('R2 (varianza explicada):', r2_score(Y_train, y_pred))
```

```python
# Realizar la grafica de dispersión y la pendiente
x = X_train.ravel() # Que los valores los reduzca a escala 10
plt.figure(figsize=(8, 5))
plt.scatter(x, Y_train, color = 'steelblue', s = 80, label = 'Datos')
plt.plot(x, y_pred, color = 'red', linewidth = 3, label = 'Pendiente')
plt.xlabel('No. Llamadas')
plt.ylabel('Copiadoras vendidas')
plt.title('Copier Sales of America')
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.show()
```
