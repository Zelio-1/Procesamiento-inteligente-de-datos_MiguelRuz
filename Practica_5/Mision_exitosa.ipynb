```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulate data for 100,000 missions
np.random.seed(42)
n_missions = 100000
EQ = np.random.choice(['Basico', 'Intermedio', 'Avanzado'], size=n_missions, p=[0.13, 0.44, 0.43]) # Equipamiento
H = np.random.choice(['Bajas', 'Medias', 'Altas'], size=n_missions, p=[0.27, 0.33, 0.40]) # Habilidades
S = np.random.choice(['Baja', 'Media', 'Alta'], size=n_missions, p=[0.24, 0.43, 0.33]) # Salud
EM = np.zeros(n_missions)
```

```python
for i in range(n_missions):
    if EQ[i] == 'Basico' and H[i] == 'Bajas' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.05, 0.95])
    elif EQ[i] == 'Basico' and H[i] == 'Bajas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.20, 0.80])
    elif EQ[i] == 'Basico' and H[i] == 'Bajas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.33, 0.67])
    elif EQ[i] == 'Basico' and H[i] == 'Medias' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.36, 0.64])
    elif EQ[i] == 'Basico' and H[i] == 'Medias' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.39, 0.61])
    elif EQ[i] == 'Basico' and H[i] == 'Medias' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.43, 0.57])
    elif EQ[i] == 'Basico' and H[i] == 'Altas' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.45, 0.55])
    elif EQ[i] == 'Basico' and H[i] == 'Altas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.47, 0.53])
    elif EQ[i] == 'Basico' and H[i] == 'Altas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.48, 0.52])
    elif EQ[i] == 'Intermedio' and H[i] == 'Bajas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.25, 0.75])
    elif EQ[i] == 'Intermedio' and H[i] == 'Bajas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.36, 0.64])
    elif EQ[i] == 'Intermedio' and H[i] == 'Medias' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.38, 0.62])
    elif EQ[i] == 'Intermedio' and H[i] == 'Medias' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.41, 0.59])
    elif EQ[i] == 'Intermedio' and H[i] == 'Medias' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.45, 0.55])
    elif EQ[i] == 'Intermedio' and H[i] == 'Altas' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.47, 0.53])
    elif EQ[i] == 'Intermedio' and H[i] == 'Altas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.49, 0.51])
    elif EQ[i] == 'Intermedio' and H[i] == 'Altas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.50, 0.50])
    elif EQ[i] == 'Avanzado' and H[i] == 'Bajas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.31, 0.69])
    elif EQ[i] == 'Avanzado' and H[i] == 'Bajas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.39, 0.61])
    elif EQ[i] == 'Avanzado' and H[i] == 'Medias' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.42, 0.58])
    elif EQ[i] == 'Avanzado' and H[i] == 'Medias' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.43, 0.57])
    elif EQ[i] == 'Avanzado' and H[i] == 'Medias' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.46, 0.54])
    elif EQ[i] == 'Avanzado' and H[i] == 'Altas' and S[i] == 'Baja':
        EM[i] = np.random.choice([1, 0], p=[0.49, 0.51])
    elif EQ[i] == 'Avanzado' and H[i] == 'Altas' and S[i] == 'Media':
        EM[i] = np.random.choice([1, 0], p=[0.51, 0.49])
    elif EQ[i] == 'Avanzado' and H[i] == 'Altas' and S[i] == 'Alta':
        EM[i] = np.random.choice([1, 0], p=[0.70, 0.30])
    else:
        EM[i] = np.random.choice([1, 0], p=[0.001, 0.999])
```

```python
# Create a DataFrame
data = pd.DataFrame({'Equipamiento': EQ, 'Habilidades': H, 'Saluid': S, 'Éxito de la Misión': EM})
```

```python
# Calculate conditional probabilities
prob_exito = data.groupby(['Equipamiento', 'Habilidades'])['Éxito de la Misión'].mean().reset_index()
```

```python
# Combine 'Equipamiento' and 'Habilidades' into a single label for the x-axis
prob_exito['Etiqueta'] = prob_exito['Equipamiento'] + ' y ' + prob_exito['Habilidades']

# Plot the probabilities
fig, ax = plt.subplots()
prob_exito.plot(kind='bar', x='Etiqueta', y='Éxito de la Misión', ax=ax, legend=False)
ax.set_ylabel('P(Éxito de la Misión = Sí)')
ax.set_title('Probabilidades Condicionales de éxito en la Misión')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
